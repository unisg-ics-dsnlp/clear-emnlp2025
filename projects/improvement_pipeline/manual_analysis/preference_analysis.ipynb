```python
def display_side_by_side(df):
    styles = [
        dict(selector="td", props=[("white-space", "pre-wrap")]),
        dict(selector="th", props=[("white-space", "pre-wrap")])
    ]
    display(HTML(df.style.set_table_styles(styles).to_html()))
```
