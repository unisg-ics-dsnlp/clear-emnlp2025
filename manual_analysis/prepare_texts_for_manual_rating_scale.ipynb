```python
texts_all = mt_en_texts + mt_de_texts + essay_texts + rev1_texts + rev2_texts + rev3_texts
texts_all
```

```python
# check if file exists
import os
if os.path.exists('preference_manual_scale/texts_all.json'):
    print("File already exists. Loading existing data.")
    with open('preference_manual_scale/texts_all.json', 'r') as f:
        texts_all = json.load(f)
else:
    print("File does not exist. Creating new data.")
    texts_all = mt_en_texts + mt_de_texts + essay_texts + rev1_texts + rev2_texts + rev3_texts
    # shuffle the texts
    import random
    random.seed(42)

    random.shuffle(texts_all)
    # save texts to a file

    with open('preference_manual_scale/texts_all.json', 'w') as f:
        json.dump(texts_all, f)
```
